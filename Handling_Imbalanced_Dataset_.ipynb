This is highly Imbalanced Dataset

0=> Legit transaction

1-> Froud transaction

```python
# separating the legit and froud transactions
legit = credit_card_data[credit_card_data.Class == 0]
froud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(froud.shape)
```

Under Sampling

Building a sample dataset similar distribution of legit and froud tarnsanctions

No of froud tansanctions -->38

```python
legit_sample = legit.sample(n=38)
```

```python
print(legit_sample)
```

```python
legit_sample.shape
```

**Concatenate the two Dataframe** 

```python
new_dataset = pd.concat([legit_sample,froud] , axis = 0)
```
