```python
#importing the dependencies
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder
```

**Label Encoding of Breast Cancer Dataset**

```python
#loading the data from pandas dataframe
Cancer_data =pd.read_csv('/content/data.csv')
```

```python
# frist 5 rows 
Cancer_data.head()
# B- Benign
# M- Malignan
```

```python
# finding the count of diffrent labels
Cancer_data['diagnosis'].value_counts()
```

```python
# load the label Encoder funtions
label_encode= LabelEncoder()
```

```python
labels = label_encode.fit_transform(Cancer_data.diagnosis)
```

```python
# appending the label to the dataframe
Cancer_data['target']= labels
```

```python
Cancer_data.head()
```

```python
Cancer_data['target'].value_counts()
```
